1. What is the difference between "Love" and "love" in NLP?

In NLP, the difference between "Love" and "love" is entirely dependent on whether our preprocessing pipeline is case-sensitive or case-insensitive.

If we haven't converted our text to lowercase (a process called Case Folding), the computer treats them as two completely different entities. The Computer’s Perspective (ASCII/Unicode) At the lowest level, computers don't see letters; they see numbers.

L has an ASCII value of 76. l has an ASCII value of 108.

Because the binary codes are different, a model will treat them as two distinct "tokens" with no relationship to each other.

2. What happens if stopwords are not removed?

Increased "Noise" and Lower Accuracy In most NLP tasks, the goal is to find the context. Stopwords like "the," "is," and "at" carry almost zero context. The Problem: A simple model might think two documents are similar just because they both use the word "the" 50 times, even if one is about Space and the other is about Cooking. The Result: Your model gets distracted by the "noise" and fails to focus on the "signal" (the important words like "Galaxy" or "Recipe").

Slower Training Times Every extra word in your dataset is a calculation the computer has to make. The Problem: If 40% of your dataset consists of "is," "and," and "the," your model is spending 40% of its training time processing words that don't change the outcome. The Result: Your training loops take much longer to finish, which is a major headache during internship deadlines or university projects.

Natural Language Generation (LLMs): If we remove "the" or "is," the AI will talk like a broken robot ("I go store" instead of "I am going to the store"). Sentiment Analysis (Sometimes): Words like "not" or "but" are often in stopword lists, but they completely change the meaning of a sentence ("I am not happy"). Part-of-Speech Tagging: We need those small words to understand the grammar and structure of a sentence.

3. Provide two real-world scenarios where removing stopwords can be harmful.

Sentiment Analysis (Negation & Context) : In sentiment analysis, stopwords like "not," "no," "but," or "never" are the "switches" that flip the entire meaning of a sentence.

Original Sentence: "The food was not good, but the service was great." After Stopword Removal: "food good service great" The Harm: The model now thinks the user had a 100% positive experience. By removing the "not," you've completely inverted the sentiment of the first half of the sentence. The word "but" also acts as a crucial structural marker that signals a shift in the speaker's opinion.

Legal or Medical Document Analysis : In fields like law or medicine, small "function words" define the boundaries of permissions, symptoms, or orders.

Scenario (Legal): "The defendant is liable for the damages." vs. "The defendant is not liable for the damages." After Stopword Removal: Both sentences become "defendant liable damages". The Harm: In a legal setting, this is a catastrophic failure. Stopwords like "is," "for," "to," and "from" define relationships between entities (who owes what to whom). Removing them turns a precise legal instruction into a vague collection of nouns.

4. Explain the difference between stemming and lemmatization

Stemming (The Rough Cutter) : Stemming works by following a fixed set of rules to chop off the ends of words (suffixes). It doesn't actually "understand" the language; it just looks for common patterns like -ing, -ed, or -es.

Lemmatization (The Professional Editor) : Lemmatization is much smarter. It looks at a dictionary (a Lexicon) and considers the context of the word—specifically its Part of Speech (POS)—to find the "Lemma" (the dictionary base form).


# Task 2: Build Advanced Preprocessing Function


```python
import re

def preprocess_text(text):
    if not text or text.strip() == "":
        return [], ""

    # Convert to lowercase
    text = text.lower()

    # Remove URLs
    text = re.sub(r'http\S+|www\S+', '', text)

    # Remove email patterns
    text = re.sub(r'\S+@\S+', '', text)

    # Remove numbers
    text = re.sub(r'\d+', '', text)

    # Handle repeated characters (soooo → soo, goooood → good)
    text = re.sub(r'(.)\1{2,}', r'\1', text)

    # Remove special characters & emojis
    text = re.sub(r'[^a-z\s]', '', text)

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # Tokenization
    tokens = text.split()

    # Remove short tokens (≤2) except 'no', 'not'
    tokens = [word for word in tokens if len(word) > 2 or word in ['no', 'not']]

    # Clean sentence
    cleaned_sentence = " ".join(tokens)

    return tokens, cleaned_sentence
```

# Task 3: Stress Testing

```python
sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

for s in sentences:
    tokens, clean = preprocess_text(s)

    print("Original Text:", s)
    print("Cleaned Tokens:", tokens)
    print("Cleaned Sentence:", clean)
    print()
```

# Task 4: Token Analytics


```python
import numpy as np

for s in sentences:
    tokens, clean = preprocess_text(s)

    total = len(tokens)
    unique = len(set(tokens))
    avg_len = np.mean([len(t) for t in tokens]) if tokens else 0

    print("Sentence:", clean)
    print("Total Tokens:", total)
    print("Unique Tokens:", unique)
    print("Average Token Length:", round(avg_len, 2))
    print()
```

# Task 5: Frequency Analysis


```python
from collections import Counter

all_tokens = []

for s in sentences:
    tokens, _ = preprocess_text(s)
    all_tokens.extend(tokens)

counter = Counter(all_tokens)

print("Top 10 most frequent words:")
print(counter.most_common(10))

print("\nTop 5 least frequent words:")
print(counter.most_common()[-5:])
```

# Task 6: Build Full Pipeline


```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, clean = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(clean)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
result = full_pipeline(sentences)
print(result)
```
